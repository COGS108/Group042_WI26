## Research Question

Does there exist a relationship between a theme park ride's attributes such as height, duration, speed and their wait time over the years 2022-2025?


## Background and Prior Work

The attractions industry is an huge sector in the field of entertainment worldwide, with the top 25 theme parks worldwide seeing 246 million visitors in 2024, a 2.4% growth from the previous year <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). With such a large impact, and millions of visitors around the world, major theme park companies are constantly trying to open world-class rides that attract more visitors to their parks. Within theme parks, some of the most popular rides are roller coasters, with over 6500 operating roller coasters worldwide, and close to 1000 in the United States alone <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2).

Wait times at theme parks, while they can be a hurdle in guest satisfaction, are often used in the amusements industry as a metric for the popularity of rides. New, heavily anticipated attractions, have attracted extremely high wait times, for example Hagrid’s Motorbike Adventure had up to 10 hour waits when it opened in 2019 <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3), and continues to attract daily waits over 2 hours, 7 years after its opening <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). 

Some of the key attributes of roller coasters that could influence their popularity are height, maximum speed, and theme, to name a few. So, it begs the question, what are the factors that lead to increased or decreased willingness to wait, which we use as a proxy for popularity, as shown by wait times in major theme park attractions.

## Hypothesis


We hypothesize that there will not be a significant relationship between higher typical thrill statistics (e.g., height, maximum speed) and increased wait time.

**What Lead Us to this Hypothesis**

While general intuition might suggest that a more conventionally thrilling ride that is taller/faster would immediately lead to more popularity, we think that it is more complicated than that. Other factors, such as IP (intellectual property), target audience (kiddie, family, thrill), and popularity of the park it is in (as measured by annual attendance), play a larger role in how people choose if they are willing to wait for an attraction.


## Data

For our ideal dataset, we would want data on approximately 600 roller coasters, which provides a large enough sample to compare different ride characteristics while still being manageable. The dataset would include variables such as average wait time (measured consistently over our selected time period), top speed, ride height, ride duration, whether the ride is associated with an intellectual property (IP), target audience (e.g., families, thrill-seekers), construction material (wood or steel), and park location.
   The data would ideally be collected from theme parks and public coaster databases, such as park wait-time tracking systems and manufacturer or industry sources. Average wait times could be gathered using historical wait-time data recorded throughout operating hours over multiple days to reduce bias from unusual crowd levels.
   Each observation in the dataset would represent one roller coaster, and the data would be stored in a tabular format, such as a CSV file or an Excel Spreadsheet. Rows would correspond to individual coasters, and columns would represent the variables listed above, with standardized units and clearly defined categories to ensure consistency and ease of analysis.
